```python
import matplotlib.pyplot as plt

# Create the figure and add some plots
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot([1, 2, 3], [2, 4, 1])
axs[0, 1].scatter([1, 2, 3], [4, 2, 3])
axs[1, 0].bar(['A', 'B', 'C'], [3, 2, 5])
axs[1, 1].hist([1, 1, 2, 3, 3, 3, 4, 5])

# Add text at the bottom of the figure
fig.text(0.5, 0.05, 'This is some text at the bottom', ha='center', fontsize=12)

# Show the plot
plt.show()
```
